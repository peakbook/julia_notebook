- Interactively plots the convex conjugate of arbitrary functions for Julia
- Original [ryotat/demo_conjugate](https://github.com/ryotat/demo_conjugate)

```julia
using Gadfly
```

```julia
function demo_conjugate(f,xx)
    x = collect(xx)
    y = map(f,x)
        
    df = diff(y)./diff(x)
    xmin = minimum(df)
    xmax = maximum(df)
    margin = (xmax-xmin)*0.1
 
    cf(y) = maximum(map(x->(x*y-f(x)),x))
    
    #cfx(y,x) = x*y - f(x) 
    #idx = [rand(1:length(x)) for i=1:3]
    #cols = [colorant"red", colorant"green", colorant"orange"]
    display(
        plot(
            Guide.title("Primal"),
            layer([f], xx[1],xx[end]),
            #[layer(x=[x[idx[i]]], y=[y[idx[i]]], Geom.point, Theme(default_color=cols[i])) for i=1:3]...             
        )
    )
    display(
        plot(
            Guide.title("Dual"),
            layer([cf],xmin-margin,xmax+margin),
            #[layer([y->cfx(y,x[idx[i]])],xmin-margin,xmax+margin,Theme(default_color=cols[i])) for i=1:3]...
        )
    )
end
```

```julia
demo_conjugate(x->log(1+exp(-x)),-5:0.1:5)
```

```julia
demo_conjugate(x->abs(x),-5:0.1:5)
```

```julia
demo_conjugate(x->(x*log(x)+(1-x)*log(1-x)),0:0.05:1)
```
