```julia
using Gadfly
using Interact
```

```julia
target(x) = sin(2*pi*x)
gendata(N) = [i for i=0.0:1.0/(N-1):1.0]
observe(xs) =  [target(x)+0.2*randn() for x in xs]
basefunc(x,n) = x^n
fitfunc(x,w) = sum([basefunc(x,m) for m=0:length(w)-1] .* w)
```

```julia
function fitting(;xs=[], ts=[], M=0, λ=0.0)
    ϕ = [basefunc(x,m) for x in xs, m=0:M]
    return inv(λ*eye(M+1)+ϕ'*ϕ)*ϕ'*ts    
end
```

```julia
function plotinteractive(N=10)
    xs=gendata(N)
    ts=observe(xs)
    @manipulate  for M=0:N-1, Regularization=[:OFF,:ON],λp=0.0:30.0
        if Regularization==:OFF
            λ=0
        else
            λ=10^(-λp)
        end
        w = fitting(xs=xs,ts=ts,M=M,λ=λ)
        
        plot(
        layer([target,x->fitfunc(x,w)],0,1),
        layer(x=xs, y=ts, Geom.point,Theme(default_color=color("blue"))),
        layer(x=xs, y=[fitfunc(x,w) for x in xs], Geom.point,Theme(default_color=color("darkorange"))),
        Coord.Cartesian(ymin=-2,ymax=2)
        )
    end
end
```

```julia
plotinteractive()
```
