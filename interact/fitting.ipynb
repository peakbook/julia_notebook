```julia
using Gadfly
using Interact
```

```julia
target(x) = sin(2*pi*x)
gendata(N) = [i for i=0.0:1.0/(N-1):1.0]
observe(xs) =  [target(x)+0.2*randn() for x in xs]
basefunc(x,n) = x^n
fitfunc(x,w) = sum([basefunc(x,i) for i=0:length(w)-1] .* w)
```

```julia
function fitting(;x=[], t=[], M=1, λ=0.0)
    ϕ = [basefunc(x[i],j) for i=1:length(x), j=0:M]
    return inv(λ*eye(M+1)+ϕ'*ϕ)*ϕ'*t    
end
```

```julia
function plotinteractive()
    N=10
    x=gendata(N)
    t=observe(x)
    @manipulate  for M=0:9, Regularization=["OFF","ON"],λp=0:30
        if Regularization=="OFF"
            λ=0
        else
            λ=10^(-λp)
        end
        w = fitting(x=x,t=t,M=M,λ=λ)
    
        fxs = [0.0:0.001:1.0]
        plot(
        layer([target,x->fitfunc(x,w)],0,1),
        layer(x=x, y=t, Geom.point,Theme(default_color=color("blue"))),
        layer(x=x, y=[fitfunc(s,w) for s in x], Geom.point,Theme(default_color=color("darkorange"))),
        Coord.Cartesian(ymin=-2,ymax=2)
        )
    end
end
```

```julia
plotinteractive()
```
